```python
# Exercício 2

num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))
num3 = float(input("Digite o terceiro número: "))


numeros = [num1, num2, num3]

maior_numero = max(numeros)
menor_numero = min(numeros)


print("-" * 30)
print(f"O maior número é: {maior_numero}")
print(f"O menor número é: {menor_numero}")
print("-" * 30)
```

```python
# Exercício 4
n = int(input("Quantos termos da Fibonacci você quer? "))

a = 1
b = 1

if n >= 1:
    print(a, end=" ")
if n >= 2:
    print(b, end=" ")

for i in range(3, n + 1):
    c = a + b
    print(c, end=" ")
    a = b
    b = c
```

```python
# Exercício 5


nome = input("Digite o nome: ")

while len(nome) < 4:
    print("Nome inválido")
    nome = input("Digite o nome novamente: ")


idade = int(input("Digite a idade: "))

while idade < 0 or idade > 150:
    print("Idade inválida")
    idade = int(input("Digite a idade novamente: "))


salario = float(input("Digite o salário: "))

while salario <= 0:
    print("Salário inválido")
    salario = float(input("Digite o salário novamente: "))


sexo = input("Digite o sexo (f ou m): ")

while sexo != "f" and sexo != "m":
    print("Sexo inválido")
    sexo = input("Digite o sexo novamente: ")


estado_civil = input("Digite o estado civil: ")

while estado_civil != "s" and estado_civil != "c" and estado_civil != "v" and estado_civil != "d":
    print("Estado civil inválido")
    estado_civil = input("Digite novamente: ")


print("Dados válidos")


```

```python
# Exercício 6

numero = int(input("Digite um número: "))

primo = True
i = 2

while i < numero:
    if numero % i == 0:
        primo = False
    i = i + 1

if numero < 2:
    primo = False

if primo:
    print("É primo")
else:
    print("Não é primo")
```

```python
# Exercício 7

numero = int(input("Digite um número: "))

fatorial = 1

for i in range(1, numero + 1):

    fatorial = fatorial * i

print(fatorial)
```

```python
# Exercício 8


lista = [5, 7, 2, 9, 4, 1, 3]

print("Lista:", lista)
print("Tamanho da lista:", len(lista))
print("Maior número:", max(lista))
print("Menor número:", min(lista))
print("Soma dos números:", sum(lista))

lista.sort()
print("Lista em ordem crescente:", lista)

print("Lista em ordem decrescente:", lista[::-1])
```
